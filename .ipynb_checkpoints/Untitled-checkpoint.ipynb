```python
# ETL_Project
import pandas as pd
import pymongo
```

```python
# Read CSV files
file1 = "Coronavirus.csv"
file2 = "Poverty Levels.csv"
```

```python
df_corona = pd.read_csv(file1)
df_corona.head()
```

```python
df_poverty = pd.read_csv(file2)
df_poverty.head()
```

```python
Complete_dataset = pd.merge(df_corona,df_poverty, how="outer", on=["State Abbr", "State Abbr"])
```

```python
Complete_dataset.head()
```

```python
Complete_dataset.dropna(subset=['Percent below poverty level'])
```

```python
Complete_dataset.fillna(0)
```

```python
Complete_dataset = Complete_dataset[Complete_dataset["State Index"] !=0]
```

```python
Complete_dataset.fillna(0)
```

```python
data_csv = Complete_dataset.to_csv("poverty_coronavirus.csv")
```

```python
import json
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.Complete_dataset
collection = db.Complete_dataset
data = json.loads(Complete_dataset.to_json(orient='records'))
collection.insert_many(data)
cursor = list(collection.find({}))
for document in cursor:
  print(document)
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
Complete_dataset.to_sql('Complete_dataset', con=engine)
```

```python
engine.execute("select * from Complete_dataset").fetchall()
```
